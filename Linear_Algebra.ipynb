Scalars, Vectors, and Matrices

Scalars are magnitude like quantities with no set direction. They are a single number, for example, temperature. 

Vectors are an array of numbers and it is defined with the operations of summation and the multiplication of a scalar. They can be refrenced as coordinates points in space with each point along a different axis. It has both magnitidude and direction in comparison to a scalar quantity. 

Matrices are 2D array of numbers with rows and columns. They represent linear maps and help describe systems of linear equations.

Multiplying matrices and vectors

If we have matrices A and B, if $A$ is *m x n* and $B$ is *n x p*, then $C$ is *m x p*.

$${C_{i, j} = \sum_k A_{i, k} B_{k, j}}$$


image.png

In order to compute the dot product between A and B we compute C as the dot product between row *i* of A and column *j* of B.

image.png

Properties of matrix multiplication: 

Distributive property: 
$${A(B +C) = AB + AC}$$

Associative property:
$${A(BC) = (AB)C}$$

However, matrix multiplication is not commutative such as: 

$${AB \neq BA}$$
